# Length of the time period

```python
df_comp.date.describe() #top generates a arbitrary values, we need to fix it
```

# from text to date

```python
df_comp.date = pd.to_datetime(df_comp.date, dayfirst=True) 
```

```python
df_comp.head()
```

```python
df_comp.date.describe()
```

# set the Index



```python
df_comp.set_index("date", inplace = True)
```

```python
df_comp.head()
```

# Setting the desired frequency

```python
df_comp = df_comp.asfreq('d')#we can set the frequency by asfreq..there 'd' means we are observing the value in daily basis
```

```python
df_comp.head()
```

```python
df_comp = df_comp.asfreq("b") #'b' means business days..means business stock exchange are off in weekend..thus if we look daily basis we are getting missing values in weekend like sat $ sunday
df_comp.head()#using 'b' tells python to expect missing values when date falls on a sat & sunday
```

# Handling Missing values

```python
df_comp.isna()
```

```python
df_comp.isna().sum() #there are 8missing values
```

```python
df_comp.spx = df_comp.spx.fillna(method = "ffill") #fill the missing values of spx with the front filling method
```

```python
df_comp.ftse = df_comp.ftse.fillna(method = "bfill") #fill the missing values of ftse with the back filling method
```

```python
df_comp.dax = df_comp.dax.fillna(value = df_comp.dax.mean())
```

```python
df_comp.nikkei = df_comp.nikkei.fillna(method = "ffill")
```

```python
df_comp.isna().sum()
```

# Add & removing the columns in dataset

```python
df_comp['market_value'] = df_comp.spx # a column named market_value is added and identical as spx
```

```python
df_comp.describe()
```

```python
del df_comp['spx'], df_comp['dax'], df_comp['ftse'], df_comp['nikkei'] #delete column 
```
